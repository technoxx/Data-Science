```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

Changing the theme of plot

```python
sns.set_theme()
```

Loading the tips dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.plot(kind='scatter', x='tip', y='size', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

Joint Distribution

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
```

```python
#iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
#titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x='class',data=titanic)
```

```python
#Bar Chart
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#Distribution Plot
sns.histplot(titanic['age'])
```

```python
from sklearn.datasets import fetch_california_housing
california_dataset = fetch_california_housing()
california_df = pd.DataFrame(california_dataset.data, columns = california_dataset.feature_names)
california_df.head()
```

```python
correlation = california_df.corr()
# Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(data=correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
